### Recursion
  - A function call itself directly or indirectly.


#### Types of recursion
  - 1. Direct : A function call itself until it reaches to base condition.
    2. Indirect : function'A' call Function 'B' and function 'B' call function 'A'. 

#### Types of recursion
  -  Direct : in it consists types
       - 1. Base condition : It is used to stop the infinite recursive call.  ex; if x==0 return 0 in below code
         2. Recursive call: The function calls recursively until it reaches base condition.

### 1. Direct Recursion

```python
# Direct recursion
def fun(x):
    if x==0:
        return 0
    x=x-1
    fun(x)
    print(x)
fun(3)
```

```python
# Direct recursion
def fun(x):
    if x==0:
        return 0
    
    fun(x-1)
    print(x)
fun(3)
```

```python
# Print 1 to n using recursion
def func(x):
    if x==0:
        return 0
    print(x)
    func(x-1)
    
func(5)
```

```python
# print factorial using recursion
def fact(x):
    if x==0 or x==1:
        return 1
    return x*fact(x-1)
print(fact(5))
```

```python
# find sum of digits in a numbers
# n=456 o/p:15
def sumOfDigits(n):
    if n==0:
        return 0
    return n%10 + sumOfDigits(n//10)
print(sumOfDigits(456))
print(sumOfDigits(15346))
```

```python
# count digits in a numbers
def sumOfDigits(n):
    if n==0:
        return 0
    return 1 + sumOfDigits(n//10)
print(sumOfDigits(456))
print(sumOfDigits(15346))
```

### Lamda Function or Anoymous Functions:
   - The function doesnt have function name, those functions called as anoymous funcs.
   - The Anoymous function can implement with 'Lambda' keyword
#### syntax:
  - lambda arguments : expression

```python
# square of number using lambda function
s=lambda x:x**2
print(s(10))
```

```python
# find square of list of integers using map and lambda
arr=[1,2,3,4,5]
print(list(map(lambda x:x**2, arr)))
```

```python
arr=[1,2,3,4,5]
print(list(filter(lambda x:x%2==0, arr)))
```

```python
# check whether no even r odd using lambda
s=lambda x:"even" if x%2==0 else "odd"
print(s(10))
```

```python
# check whether no even r odd using lambda
n=int(input())
s=lambda x:"even" if x%2==0 else "odd"
print(s(n))
```

```python
arr=[1,2,3,4,5]
# 1-odd
# 2- even
arr=[1,2,3,4,5]
print(list(map(lambda x:f"{x} - even" if x%2==0 else f"{x} - odd",arr)))
```

```python
# count the frequency of each character
string=input()
count={}
for char in string:
    if char in count:
        count[char] +=1
    else:
        count[char]=1
print(count)
```

```python
# count the frequency of each character
string=input()
count={}
for char in string:
    count[char] = count.get(char,0)+1
print(count)
```
