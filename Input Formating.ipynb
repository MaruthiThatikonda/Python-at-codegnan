#### Input Formating
 1. Integer reading from user

```python
# reading integer from user 
num=int(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# reading float value from user
val=float(input())
print(type(val))
print(val)
```

```python
# complex reading from user
val=complex(int(input()),int( input()))
print(type(val))
print(val)
```

```python
# boolean reading from user
val=bool(int(input()))
print(type(val))
print(val)
```

```python
# string value reading from user
val=input("enter name:")
print(type(val))
print(val)
```

```python
# reading List of integers from user
lst =input().split(',')
print(lst)
```

### map(): This function applies functionality on each element is iterable


```python
lst =map(int,input().split())
print(lst)
```

```python
lst =list(map(int,input().split()))
print(lst)
```

```python
# reading list of string from user
v=list(map(str,input().split()))
print(v)
print(type(v))
```

```python
# approach 2
lst=input().split(',')
print(lst)
```

```python
# tuple of integers from user
v=tuple(map(int,input().split()))
print(v)
```

```python
# set of integers from user
m=set(map(int,input().split()))
print(m)
```
