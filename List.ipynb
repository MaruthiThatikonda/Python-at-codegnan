### List Methods 
- Item adding methods
  - append(item) : It adds item at end of list
  - extend (iterable) : It adds the all iterator elements at end of list
  - insert(pos, item) : It adds item at specified position 

```python
lst= [10,20,30,40]
lst.append(80)
print(lst)
```

```python
lst= [10,20,30,40]
lst.extend([1,2,3])
print(lst)
lst.append([100,200])
print(lst)
```

```python
lst= [10,20,30,40]
lst.insert(0,250)
print(lst)
```

### Item removing methods
  - remove(item): It Removes the first occurence Item form the list if it is exists in list
    otherwise 'ValueError'
  - pop(pos):It removes and return the specified position item if position is passed,
    otherwise it remove and returns the last item from list
  - clear(): It removes all elements from list 

```python
lst = [10,20,30,20,50]
lst.remove(20)
print(lst)
lst.remove(100) #ValueError: list.remove(x): x not in list
print(lst)
```

```python
lst=[10,20,30,20,40,50,20]
val = lst.pop()
print(val)
print(lst)
val2=lst.pop(0) #It removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```
