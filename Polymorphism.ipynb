### Polymorphism:
  - In python polymorphism the poly means-> 'many' and morphism means-> 'forms'
  - In polymorphism single function it shows different behaviours based on input context.(values)
  - In python polymorphism can achieve through the following ways:
     - 1.Through operators
     - 2.Polymorphism through functions
     - 3.Polymorphism through methods
       - 3.1.Runtime polymorphism (method overriding).
       - 3.2.Compile time polymorphism (method overloading).

#### 1.Polymorphism through operators
- Here the single operator it shows different actions (or) behaviours based on input context. (or) input values.
- EX:
  - '+' operator it perform addition operation when inputs are integers. as well as it performs string concatination when inputs are strings.
  - EX: 3+8 -> 8 but operation is addition
       - 'code'+'gnan'->'codegnan' operation is string concatnation
  -  '*' operator it perform multiplication operation when inputs are integers. as well as it performs string concatination when inputs are strings.
  - EX: 3*8 -> 24 but operation is multiplication
  - 'gnan' *3->'gnan gnan gnan' operation is string repitation


#### 2.Polymorphism through functions
- Here a function shows different actions based on input context.
- ex:
- len('hero') -> string length
- len([10,20,30])-> list length

#### 3.Polymorphism through the Methods
 - ##### 1. Method overriding (runtime polymorphism).
    - A method defined in the child class that is already exist in parent class with the same method.
    - whenever if we call that method through the child class object the child class method overrides the parent class method.
    - This is called as method overriding.
    - ex:
       - class A:
         - def show() # same method 
       - class B:
         - def show() # same method 
       - b=B()
       - b.show() # calling same method from child class only

```python
# method overriding
class person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self): #instance method
        return f"My name is {self.name} and I am {self.age} years old Gender {self.gender}"

class student(person):
    def __init__(self,stdid, name, age, gender,branch):
        super().__init__(name,age,gender)
        self.stdid=stdid
        self.branch=branch
    def details(self):
        return f"My name is {self.name} and I am belongs to {self.branch}"

s=student(101,"Mahi",45,"male","pfs")
print(s.details())  # it will executes of child class method it is called child class method override the parent class method.    
```

#### 2.Method overloading
 - A class contains multiple methods with same name with different parameters list, Int, count last defined method overrides the previous defined method.This is called as Method overloading.
 - Directly python doesn't supports the method overloading but we can achieve same functionality through the default arguments, variable length, positional arguments (or) variable length keyword arguments.
 - EX:
    - class addition:
       - def add(a,b):
         - return a+b
       - def  add(a,b,c):
         - return a+b+c
       - def add(a,b,c,d):
          - return a+b+c+d
- in this case if we pass the arguments like a,b=2,5 in 1st it will show error because in this method overloadding last function will print no previous will work

```python
# method overloading
class Math:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d

b=Math() 
#print(b.add(2,3)) #missing 2 required positional arguments: 'c' and 'd'
#print(b.add(2,3,4)) #missing 1 required positional argument: 'd'
print(b.add(1,2,3,4)) # only print of last fun based on arguments 
```

```python
# method overloading
# directly python doen't support the method overloading
# but we can achieve same fun through the default arguments or arbitary arguments
class Math:
    def add(self,a,b,c=0,d=0):   # here c=0 d=0 are default arguments.
        return a+b
    

b=Math() 
print(b.add(2,3)) 
print(b.add(2,3,4))
print(b.add(1,2,3,4)) 
```

```python
# Magic methods or Special methods
class Point:
    def __init__(self,x):
        self.x=x
p1=Point(5)
print(p1)
```

```python
# __str__() magic methods
class Point:
    def __init__(self,x):
        self.x=x
    def __str__(self): # string magic method
        return f"This is point class and point value is {self.x}"
p1=Point(5)
print(p1)
```

```python
# __str__() magic methods
class Point:
    def __init__(self,x):
        self.x=x
    def __str__(self): # string magic method
        return f"This is point class and point value is {self.x}"
    def __add__(self,other):
        return self.x + other.x
p1=Point(5)
print(p1)
p2=Point(3)
print(p1.x+p2.x)
# or 
print(p1+p2)
# or
p3=p1+p2
print(p3)
```

```python
class Point:
    def __init__(self,x):
        self.x=x
    def __add__(self,other):
        return self.x + other.x
    def __sub__(self,other):
        return self.x - other.x
    def __mul__(self,other):
        return self.x * other.x
    def __truediv__(self,other): # truediv for division
        return self.x /other.x
    def __mod__(self,other):
        return self.x % other.x
p1=Point(5)
print(p1)
p2=Point(4)
print(p2)
print(p1+p2)
print(p1-p2)
print(p1* p2)
print(p1/p2)
print(p1%p2)
```
