### LOOPS
  - Perform same operation repeadtly again and again, this is called as loops.

#### Types of loops
 - 1. for loop
 - 2. while loop

##### for Loop:
   - syntax: - for iterator in iterable:
     - ##### block of code
            - else:
       - ###### block of code

```python
# check no is even or odd
lst=list(map(int,input().split()))
for i in lst:
    if i%2==0:
        print(f'{i} is even')
    else:
        print(f'{i} is odd')
else:
    print("Program done")
```

```python
# check no is even or odd if even make square if odd add +1
lst=list(map(int,input().split()))
for i in lst:
    if i%2==0:
        print(i**2)
    else:
        print(i+1)
else:
    print("Program done")
```

### range function syntax
       - range(start,end,step)
parameter is end='' to reprsent in ans in 1 single line

```python
# print 1 to n numbers 
n=int(input())
for i in range(1,n+1):# n+1 means it will print upto n value
    print(i,end=' ')# end=' ' it will give space and print ans in 1single line
```

#### while loop():
 syntax :
 - initilization
 - while condition:
   - #block of code
   - #increment/decrement
 - else:(optional)
   - #block of code
  


```python
# print 1 to n numbers using while loop
n=int(input())
i=1
while i<=n:
    print(i,end=' ')
    i+=1
```

#### The loop body executes until condition evalutes False:
 - In while loop manual increment/decrement
 - If you don't know the total no of iterations, we will use while loop.

```python
lst=[10,20,30,40,50]
tar=40
n=len(lst)
for i in range(n):
    if lst[i]==tar:
        print(i)
        break
else:
    print("-1")
        
```

```python
s="maruthi"
tar="a"
n=len(s)
for i in range(n):
    if s[i]==tar:
        print(i)
        break
else:
    print("-1")
```

#### Loop control statements
 #### 1.Continue: 
   - After executing the 'continue' keyword execution immediately it goes to next iteration and skips the current iteration after the continue keyword statement

```python
# Loop Control statement
# 1.Continue
for i in range(1,11):
    if i%2==0:
        continue # skip the ans and goes to next iteration.
    print(i)
```

#### 2.Break:
  - After executing of 'breaks' keyword immediately execution goes to out of loop.
  - That means loop will terminate
  - The loops else block executes when break keyword not executed.

```python
lst=list(map(int,input().split()))
tar=int(input())
n=len(lst)
for i in range(n):
    if lst[i]==tar:
        print(i)
        break
else:
    print("-1")
```

```python
# Factors of a number to print in list form
n=int(input())
factors=list()
for i in range(1,n+1):
    if n%i==0:
        factors.append(i)
print(factors)
```

```python
# Prime number or not
n=int(input())
factors=list()
for i in range(1,n+1):
    if n%i==0:
        factors.append(i)
if len(factors)==2:
    print(f'{n} is prime')
else:
    print(f'{n} is not prime')
```

```python
# prime r not
n=int(input())
for i in range(2,n//2+1):
    if n%i==0:
        print(f'{n} is not prime')
        break
else:
    print(f'{n} is prime')
```

#### Nested Loop:
    - A loop contains one another loop
     - Ex: for i in range(1,10):
            - for j in range(1,10)
                - print(i,j)

```python
#### 
n=int(input())
for j in range(2,n):
    for i in range(2,n//2+1):
        if n%i==0:
            print(f'{n} is not prime')
            break  
else:
    print(f'{n} is prime')
```

#### Patterns
  - i indicates rows
  - j indicates columns
  - end=' ' print in seperate new line 


```python
n=5
for i in range(n):
    for j in range(n):
        print("*",end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print(i+1,end=" ") # rows are same so i+1
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        print(j+1,end=" ") # colums are same so j+1
    print()
```

```python
n=4
for i in range(n):
    for j in range(n):
        print(n*i+j+1,end=" ")
    print()
```
