### Conditional Statements
 - They perform of conditional statements is to make decisions.
 #### Types
  - 1. if
    2. if-else
    3. if-elif-else
    4. Nested-if

#### 1. if statement : The block code is executes when if condition evalutes to True, otherwise Nothing
##### syntax:
    if condition:
        block of code

```python
x=10
if x==0:
    print("Value is 0")
print("program done")
```

#### 2.if-else statement: The else block code executes when condition evalutes to False, otherwise it executes if block code
##### syntax
  - if condition:
     - block of code
   - else:
     - block of code 
     

```python
x=10
if x==0:
    print("Value is 0")
else:
    print("Value is 10")
```

```python
# prbl: check whether the number is divisible with 3 or not
x=int(input("Enter number:"))
if (x%3==0):
    print("Number is divisible by 3")
else:
    print("Number is not divisible by 3")
```

```python
## 3.genereate bill if total bill is morethan 500 then give 10% discount otherwise 0% discount
Totalbill=int(input("Enter the bill amount:"))
if (Totalbill>500):
    discount=Totalbill*10/100
    discount_bill=Totalbill-discount
    print(f"{ discount_bill} is your bill amount by 10% discount")
else:
    print(f"{ Totalbill}is your bill amount without discount")
```

```python
## 4.check weather the product stock is avaliable or not (read product name and stock)
product=(input("Enter The product:"))
stock=int(input())
if stock>0:
    print(f"Currently {product} : {stock} items avaliable")
else:
    print(f" Currently {product}is out of stock")
```

```python
## 5.Question
products={'pen':5,'book':20,'pencil':3}
item=input().lower()
quantity=int(input())
if item in products:
    total_amount=products.get(item)* quantity
    print(f'Total amount of {quantity} {item} is: {total_amount}')
else:
    print(f"Item not avaliable")
```

#### 3.if-elif-else: 
   - The if block executes when if condition evalutes to True, otherwise it sequentially checks with elif condition if any condition is True then it executes respective blocks otherwise else block will executes
  - If any block is executed immediately execution will goes to Main block (or) Out of blocks.
##### syntax
     - if condition1:
       -  block1
    - elif condition2: 
        - block2
    - elif condition3: 
        - block3
    - else:
        - block of code

```python
##### largest of 3no's
x=int(input("Enter 1's no:"))
y=int(input("Enter 2's no:"))
z=int(input("Enter 3's no:"))
if(x>=y and x>=z):
      print("X is larger no")
elif(y>=x and y>=z):
      print("Y is larger no")
else:
    print("Z is larger no")
```

```python
# check whether no is positive even negative even positive odd negative odd or 0
n=int(input())
if n%2==0 and n>=0:
    print("Positive even")
elif n%2==0 and n<0:
    print("Negative even")
elif n%2!=0 and n>0:
    print("Positive odd")
elif n%2!=0 and n<0:
    print("Negative odd")
else:
    print("Invalid number")
```

#### Nested-if:
##### syntax:
   - if condition1:
      - if condition2:
        -  #block1
     -  else:
         - #block2
   - else:
      - if condition3:
        -  #block3
      - else:
         - #block4

```python
# nested-if
# check whether no is positive even negative even positive odd negative odd or 0
n=int(input())
if n%2==0:
    if n>=0:
        print(f"{n} +ve even") 
    else:
        print(f"{n} -ve even")
elif n%2!=0:
    if n>0:
        print(f"{n} +ve odd")
    else:
        print(f"{n} -ve odd")
else:
    print(f"{n} invalid")
```

```python
##### largest of 3no's
n1=int(input())
n2=int(input())
n3=int(input())
if n1>=n2:
    if n1>=n3:
        print("n1 is larger")
    else:
        print("n3 is larger")
else:
    if n2>=n3:
        print("n2 is larger")
    else:
        print("n3 is larger")
    
```
