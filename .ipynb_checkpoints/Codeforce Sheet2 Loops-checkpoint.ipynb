```python
# A. 1 to N
n=int(input())
for i in range(1,n+1):
    print(i)
```

```python
#B. Even Numbers
N=int(input())
if n<2:
        print("-1")
else:
    for i in range(2,N+1,2):
        print(i)
```

```python
#C. Even, Odd, Positive and Negative
N = int(input())
arr = list(map(int, input().split()))

even = 0
odd = 0
positive = 0
negative = 0

for X in arr:
    if X % 2 == 0:
        even += 1
    else:
        odd += 1

    if X > 0:
        positive += 1
    elif X < 0:
        negative += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)
```

```python
#D. Fixed Password
while True:
    x = int(input())

    if x == 1999:
        print("Correct")
        break
    else:
        print("Wrong")
```

```python
#E.Max of number 
n = int(input())
arr = list(map(int, input().split()))

maximum = arr[0]

for i in arr:
    if i > maximum:
        maximum = i

print(maximum)
```

```python
# F.Multiplication of table
n = int(input())

for i in range(1, 13):
    print(f"{n} * {i} = {n * i}")
```

```python
# G.Factorial
t = int(input())

for _ in range(t):
    n = int(input())

    fact = 1
    for i in range(1, n + 1):
        fact *= i

    print(fact)
```

```python
# H.One Prime
n = int(input())

prime = True

for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
        prime = False
        break

if prime:
    print("YES")
else:
    print("NO")
```
