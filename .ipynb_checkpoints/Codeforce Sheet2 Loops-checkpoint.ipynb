```python
#C. Even, Odd, Positive and Negative
N = int(input())
arr = list(map(int, input().split()))

even = 0
odd = 0
positive = 0
negative = 0

for X in arr:
    if X % 2 == 0:
        even += 1
    else:
        odd += 1

    if X > 0:
        positive += 1
    elif X < 0:
        negative += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)
```

```python
#E.Max of number 
n = int(input())
arr = list(map(int, input().split()))

maximum = arr[0]

for i in arr:
    if i > maximum:
        maximum = i

print(maximum)
```

```python
# F.Multiplication of table
n = int(input())

for i in range(1, 13):
    print(f"{n} * {i} = {n * i}")
```

```python
# G.Factorial
t = int(input())

for _ in range(t):
    n = int(input())

    fact = 1
    for i in range(1, n + 1):
        fact *= i

    print(fact)
```

```python
# H.One Prime
n = int(input())

prime = True

for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
        prime = False
        break

if prime:
    print("YES")
else:
    print("NO")
```

```python
# I. Palindrome
n = input()

rev = n[::-1]
print(int(rev))  # Removes leading zeros

if n == rev:
    print("YES")
else:
    print("NO")
```

```python
# J. Primes from 1 to n
n = int(input())

for i in range(2, n + 1):
    prime = True
    for j in range(2, int(i ** 0.5) + 1):
        if i % j == 0:
            prime = False
            break
    if prime:
        print(i, end=" ")
```

```python
# K. Divisors
n = int(input())

for i in range(1, n + 1):
    if n % i == 0:
        print(i)
```

```python
# L. GCD
a, b = map(int, input().split())

while b != 0:
    a, b = b, a % b

print(a)
```

```python
# M. Lucky Numbers
a, b = map(int, input().split())

found = False

for i in range(a, b + 1):
    s = str(i)
    lucky = True

    for ch in s:
        if ch != '4' and ch != '7':
            lucky = False
            break

    if lucky:
        print(i, end=" ")
        found = True

if not found:
    print(-1)
```

```python
# O. Pyramid
N=int(input())
for i in range(N):
    for j in range(N):
        if i>=j:
            print("*",end="")
        else:
            print(" ",end=" ")
    print()
```

```python
#Q. Digits
t = int(input())

for _ in range(t):
    n = int(input())

    if n == 0:
        print(0)
    else:
        while n > 0:
            print(n % 10, end=" ")
            n //= 10
        print()
```

```python
#R. Sequence of Numbers and Sum
while True:
    n, m = map(int, input().split())

    if n <= 0 or m <= 0:
        break

    if n > m:
        n, m = m, n

    total = 0

    for i in range(n, m + 1):
        print(i, end=" ")
        total += i

    print(f"sum ={total}")
```

```python
# S. Sum of Consecutive Odd Numbers
t = int(input())

for _ in range(t):
    x, y = map(int, input().split())

    if x > y:
        x, y = y, x

    total = 0

    for i in range(x + 1, y):
        if i % 2 != 0:
            total += i

    print(total)
```

```python
#U. Some Sums
n, a, b = map(int, input().split())

ans = 0

for i in range(1, n + 1):
    digit_sum = 0
    temp = i

    while temp > 0:
        digit_sum += temp % 10
        temp //= 10

    if a <= digit_sum <= b:
        ans += i

print(ans)
```

```python
#V. PUM
n = int(input())

x = 1

for i in range(n):
    print(x, x + 1, x + 2, "PUM")
    x += 4
```

```python
#W. Shape3
n = int(input())

# Upper half
for i in range(n):
    spaces = n - i - 1
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)

# Lower half
for i in range(n - 1, -1, -1):
    spaces = n - i - 1
    stars = 2 * i + 1
    print(" " * spaces + "*" * stars)
```

```python
#X. Convert To Decimal 2
t = int(input())

for _ in range(t):
    n = int(input())

    ones = bin(n).count('1')
    ans = (1 << ones) - 1

    print(ans)
```

```python
#Y. Easy Fibonacci
n = int(input())

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#Z. Three Numbers
k, s = map(int, input().split())

count = 0

for x in range(k + 1):
    for y in range(k + 1):
        z = s - x - y
        if 0 <= z <= k:
            count += 1

print(count)
```
