### TYPE CONVERSION
- TYPE CONVERSION: Converting one datatype format into another datatype format.
This is called as Type Conversion
- This are 2 types:
   - 1.implicit type conversion
   - 2.Explict type conversion(Type casting)

#### 1.Implicit type conversion 
  - The python interpretor automatically converts the datatype bcz it returns the data loss
  - Implicit type conversion can happens in arithmetic operator

#### 2.Explicit type conversion :(Type casting)
  - Here the type conversion happens using python built in datatype functions.

#### Python Built in datatype functions
  1.Integer- int()
  2.Float value- float()
  3.Complex value- complex()
  4.Boolean- bool()
  5.String - str()
  6.List- list()
  7.Tuple - tuple()
  8.Set- set()
  9.Dictionary - dict()

```python
# integer to float, boolean, string
num=10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
```

```python
# float to int ,bool, string
val=10.2556
num= int(val)
print(type(num),num)
print(bool(val))
str_val=str(val)
print(type(str_val),str_val)
print(str_val[:4])
print(str_val[1])
print(str_val[-1])
```

```python
# string to int,float,bool,list,tuple,set
# if string value is number
s='123456'
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# string to int,float,bool,list,tuple,set
# if string value is float 
s='123456.12305'
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#list to str,tup,set
s=[1,3,4,'hi']
print(str(s))
print(tuple(s))
print(set(s))
```

```python
# Tuple to str,list,set
t=(1,4,9,'hello')
print(str(t))
print(list(t))
print(set(t))
```

```python
#Set to str,list,tuple
s={1,5,8,'msd'}
print(str(s))
print(list(s))
print(tuple(s))
```

```python
#dict to str,list,set,tuple
d={"Name":"Shreyas","Course":"Python","Batch":57,"Age":22}
print(str(d))
print(list(d))
print(set(d))
print(tuple(d))
```
