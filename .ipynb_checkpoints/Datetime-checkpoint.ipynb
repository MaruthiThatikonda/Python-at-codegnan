```python
from datetime import datetime
current_timestamp = datetime.now() # to get the current timestamp.
print(current_timestamp)
today=current_timestamp.date() # to extract date from current timestamp.
print(today)
curr_time=current_timestamp.time() # It extracts the time from timestamp.
print(curr_time)
# extract day,month,year
day=today.day
month=today.month
year=today.year
print(day,month,year)
# extract hour,minute,and seconds from time
print(curr_time.hour)
print(curr_time.minute)
print(curr_time.second)
```

```python
from datetime import datetime
current_timestamp = datetime.now() # to get the current timestamp.
print(current_timestamp)
today=current_timestamp.date() # to extract date from current timestamp.
print(today)
curr_time=current_timestamp.time() # It extracts the time from timestamp.
print(curr_time)
# extract day,month,year
day=today.day
month=today.month
year=today.year
print(day,month,year)
# extract hour,minute,and seconds from time
print(curr_time.hour)
print(curr_time.minute)
print(curr_time.second)
# change date format
print(today.strftime("%d / %m / %y"))
print(curr_time.strftime("%I: %M %p"))
```

```python
# Timer implementation
import time
hour_counter = minute_counter=second_counter=0
while True:
    print(f"{hour_counter} : {minute_counter}: {second_counter}")
    time.sleep(1) # It pauses the execution for 1 sec
    second_counter +=1
    if second_counter==60:
        minute_counter+=1
        second_counter=0 # reset seconds counter to 0 
        if minute_counter==60:
            hour_counter +=1
            minute_counter=0 # reset mincounter to 0
            if hour_counter ==24:
                hour_counter=0
```

```python
# Timer implementation
import time
from IPython.display import clear_output
hour_counter = minute_counter=second_counter=0
while True:
    print(f"{hour_counter} : {minute_counter}: {second_counter}")
    time.sleep(1) # It pauses the execution for 1 sec
    second_counter +=1
    if second_counter==60:
        minute_counter+=1
        second_counter=0 # reset seconds counter to 0 
        if minute_counter==60:
            hour_counter +=1
            minute_counter=0 # reset mincounter to 0
            if hour_counter ==24:
                hour_counter=0
    clear_output(wait=True)
```
