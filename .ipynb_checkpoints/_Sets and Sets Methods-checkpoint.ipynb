### SETS
 - Sets is unordered collection of different datatype elements sets are mutable and defines with '{ }'

## characterstics 
 - Set is a unordered collection
 - set elements are unique
 - set doesnot supports indexing
 - set are mutable
 - set allows hashable datatype as elements
 - set elements can be used as dictionary keys


```python
#set unorder collection
# set return unique elements
s={1,2,3,2,3,'hi','a',2.02,(10,20,30)} 
print(s)
#print(s[0]) # TypeError: 'set' object is not support indexing 
```

```python
print({10,6,3,5})
print(hash(10),hash(3))
```

```python
# List can't be used as set element
s={1,2,3,2,3,'hi','a',2.02,(10,20,30)}
print(s)
```

```python
# empty set representation
s1={ }
s2=set()
print(type(s1), type(s2))
```

### Set Methods
 -   ### Element adding method
 -  add(ele): It adds the element to set if element not in set,otherwise nothing

```python
# Element adding method
# add(ele)
s={1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
print(s)
```

- ### Element removing methods
 - remove(item):It removes the item from set  if it exists in set otherwise,'KeyError'
 - discard(item):It used to removes the item if it exists in set otherwise nothing

```python
s={1,2,3,4,5}
s.remove(5)
print(s)
# s.remove(100) # KeyError
s.discard(3)
print(s)
s.discard(100) # It returns nothing
print(s)
```

### Sets Operations
 - Union(set): It returns the all the unique elements from all sets
 - intersection(set): It returns common elements from all sets
 - difference(set): It returns a set elements those are not in another sets
 - symmetric_difference(set): It returns non common elemts bw all sets

```python
# union
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
u=s1.union(s2,s3)
print(u)
res=s1|s2|s3
print(res)
```

```python
# intersection
s1={1,3,4,7}
s2={4,1,6,7}
s3={1,2,4,9}
u=s1.intersection(s2,s3)
print(u)
res=s1&s2&s3
print(res)
```

```python
# difference
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
u=s1.difference(s2,s3)
print(u)
res=s1-s2-s3
print(res)
```

```python
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
print((s1|s2)-s3)
```

```python
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
u=s1.symmetric_difference(s2)
print(u)
res=s1^s2
print(res)

```

```python
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
u=s1.symmetric_difference(s2)# in method use only 2 sets
print(u)
res=s1^s2^s3 # in symbols use original sets
print(res)

```

### Subset
- issubset(set)
### Superset
- issuperset(set)
### Disjoint
- isdisjoint(set)
   - All returns True or False (boolean values)

```python
s1={5,7}
s2={4,5,6,7}
m=s1.issubset(s2)
print(m)
```

```python
s1={5,7}
s2={4,5,6,7}
m=s1.issuperset(s2)
print(m)
```

```python
s1={1,3,}
s2={4,5,6,7}
n=s1.isdisjoint(s2)
print(n)
```

```python
s1={1,2,3}
s2={1,2,3,4}
s3={5}
print(s1.issubset(s2))
print(s2.issubset(s1))
print(s2.issuperset(s1))
print(s3.isdisjoint(s1))
```
