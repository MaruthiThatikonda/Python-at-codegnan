#### Built in functions
  - The functions defination is provided by the python software those functions called it as Built in functions.

```python
# round(val,pos):
print(round(10.2584,2))
print(round(10.2542,2))
```

```python
# chr() : This functions returns the character based on ASCII value
# ord() : It returns the ASCII value based on the character
print(chr(65))
print(ord("A"))
print(ord('a'))
print(ord('1'))
```

```python
# abs() : It returns the abosloute value no negatives, it convert neg-ve number to +ve number
print(abs(10-20))
```

```python
# Find the difference bw two strings
s1="CODE"
s2="GNAN"
n=len(s1)
s=""
for i in range(n):
    diff= abs(ord(s1[i])-ord(s2[i]))
    if diff==0:
        s+=s1[i]
    else:
        s+=chr(64+diff)
print(s)
```

```python
# Find the difference bw two strings
s1="CODEA"
s2="GNANA"
n=len(s1)
s=""
for i in range(n):
    diff= abs(ord(s1[i])-ord(s2[i]))
    if diff==0:
        s+=s1[i]
    else:
        s+=chr(64+diff)
print(s)
```

```python
# Find the bigger and smaller characters from s1 and s2
# Compare each character in both strings and if s1[i]> s2[i] and s1[i] to bigger and s2[i] to bigger 
# if both same leave it
s1="CODE"
s2="GNAN"
n=len(s1)
a=""
b=""
for i in range(n):
    if((ord(s1[i]))>(ord(s2[i]))):
        a+=s1[i]
        b+=s2[i]
    elif((ord(s2[i]))>(ord(s1[i]))):
        a+=s2[i]
        b+=s1[i]
print(a)
print(b)
```

```python
# Find the bigger and smaller characters from s1 and s2
# Compare each character in both strings and if s1[i]> s2[i] and s1[i] to bigger and s2[i] to bigger 
# if both same leave it
s1,s2=map(str,input().upper().split())
n=len(s1)
big = small=""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i]>s2[i]:
            big+=s1[i]
            small+=s2[i]
        else:
            big+=s2[i]
            small+=s1[i]
print(big)
print(small)
```

```python
# map(fun,iterable): It applies the functionality on each element in iterable
# filter(fun,iterable): It filters the itables, it returns the value which is not None 
# find square of list of integers
def squere(x):
    return x**2
lst=[1,2,3,4,5]
print(list(map(squere,lst)))
```

```python
# filter the even numbers from list of integers
def even(x):
    if x%2==0:
        return x
        
arr=[1,2,3,4,5,6,7]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
def squere(t):
    x,y=t
    return x+y
lst=[(1,2),(3,4)]
print(list(map(squere,lst)))
```

```python
# zip(iterale1,iteable2,.....iterableN):
# It returns list of tuples of 
l1=['a','b','c']
l2=[1,2,3,4,5]
print(list(zip(l1,l2)))
```

```python
# zip(iterale1,iteable2,.....iterableN):
# It returns list of tuples of tuples as for each tuple is generated a each position element from all iterable
names=['ravi','srinu','shiva']
age=[20,22,24]
print(list(zip(names,age)))
```

```python
# enumerate(itable): It returns the list of tuple of index and value
arr=[10,20,30]
print(list(enumerate(arr)))
```

```python
# find the target element position in list
arr=[10,20,30,40,50]
target=40
for ind,val in enumerate(arr):
    if target==val:
        print(ind)
        break
else:
    print(-1)    
```

```python
# any(iterable): It returns True if any one element in the iterable boolean value is true 
arr=[1,2,4,0,0,0]
print(any(arr))
print(any([0,0,0]))
# all(iterable): It returns True when all  iterble elements boolean value is True
print(all([1,2,3,4]))
print(all([True,False,False]))
```

```python
# isinstance(obj,class_name): It checks weather the object is belongs to particular class
print(isinstance(10,int))
print(isinstance(12.25,int))
```

```python
# sorted(): It returns the new sorted object, by default it returns the descending order
# reverse(): It returns the reverse of an object value
arr=[10,2,3,4,3,5,6]
res=sorted(arr)
print(res)
```

```python
# sorted(): It returns the new sorted object, by default it returns the descending order
# reversed(): It returns the reverse of an object value
arr=[10,2,3,4,3,5,6]
res=sorted(arr,reverse=True)
print(res)
print(list(reversed(arr)))
print("".join(list(reversed("Hello"))))
```

```python
arr=["Hello",'hi','apple','banana']
sorted(arr,key=len)
```

```python
# count the frequence of characters in string
string="babu"
for char in string:
    print(f'{char}-{string.count(char)}')
```

```python
string="babu"
dic=dict()
for i in string:
    dic[i]=string.count(i)
print(dic)
print(list(dic.values()))
```
