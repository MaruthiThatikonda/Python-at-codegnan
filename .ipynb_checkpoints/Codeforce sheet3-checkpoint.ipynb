```python
# 2.
def printNumbers(n):
    for i in range(1,n+1):
        print(i,end=" ")
m=int(input())
printNumbers(m)
```

```python

n=int(input())
res=""
while n>0:
    rem=n%2
    res=str(rem)+res
    n=n//2
print(res)
```

```python
# reverse a string without slicing
string=input()
rev=" "
for char in string:
    rev=char +rev
print(rev)
```

```python
# find the binary value and check whether it is binary value is palindrome or not
n=int(input())
res=""
while n>0:
    rem=n%2
    res=str(rem)+res
    n=n//2
print(res)
rev=""
for char in res:
    rev=char+rev
print(rev)
print("Palindrome" if res== rev else "Not palindrome")
```
