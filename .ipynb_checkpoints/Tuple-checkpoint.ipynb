### Tuple
 - Tuple is ordered collection of different data type elements .Tuple is immutable and defines with '()'
 - Characteristics:
    - Tuple is oredered collection
    - Tuple is immutable
    - Tuple is allows indexing slicing
    - Tuple is Heterogenous datatype(allows diff datatypes of elts)
    - Tuple allows duplicate items.
 - Ex: t=(1,2,3.5,'Hello',[10,20,30])

### Differnce bw tuple and list in notes

### Nested indexing 


```python
t=[1,2,[4,5,6],[7,8],9,1]
print(t[2][2])
```

```python
#immutable 
#After operation it doesnot change the length of object and doen't change the remaining elts position
t= (1,2,3,[10,30,40],23)
t[3].append(100)
print(t)
```

```python
# empty tuple representation 
#Single elt repesentation 
t1=()
t2= tuple()
print(type(t1),type(t2))
```

```python
# single element representation 
t1=(10)
t2=(10,)
l1=[10]
print(type(t1),type(t2),type(l1))
```

```python
(0,) *10
```

```python
# Tuple operations
# Tuple concatination 
t1= (1,2,3,23)
t2=(4,5,8,7)
s=t1+t2
print(s)
# Tuple Repetation
t=(4,5,8,7)
print(t*2)
```

```python
#Tuple methods
# index()
t2=(4,5,8,7)
print(t2.index(4))
# count()
t2=(4,5,8,7,8,8)
print(t2.count(8))
```

### Builtin funcs()

```python
# Builtin Functions
t = (1,2,3,4,5)
print(len(t))
print(min(t))
print(max(t))
print(sum(t))
print(sum(t) / len(t)) # Average
```
