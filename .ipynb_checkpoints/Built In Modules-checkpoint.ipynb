```python
import math
print(math.ceil(4.25)) # It returns upperbound integer value
print(math.floor(4.25)) # It returns Lowerbound integer value
print(math.factorial(5)) # It returns the factorial of a number
print(math.gcd(10,6,4)) # It returns the Greatest common Divisor
print(math.lcm(10,6,4)) # It returns the Least common multipler
print(math.pow(10,3)) # pow(a,b) It returns the Float type of a raise to b
print(math.sqrt(12)) # It returns the squareroot value of a number 
print(math.log10(4)) # To find the log10 values 
print(math.sin(45)) # It returns sin values
```

```python
import random
print(random.random()) # This function Generates the random number between (0 to 1)
print(random.randint(1,6)) # This function Generates the random number between (1 to 6)
print(random.randint(1000,9999)) # 4 digit OTP generates
print(random.choice(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'])) # This function returns the random value from sequence
print(random.choices(['A',1,2,3,4,5,6,7,8,9,'J','Q','K'],k=13)) # This function used in Online Rummy game
arr=['A',1,2,3,4,5,6,7,8,9,'J','Q','K']
random.shuffle(arr)
print(arr)
```

```python
# Who will reach Top[100] in less steps by Dices
count=rolls=0
print("Welcome to the who will Reach the Top In less Steps Game")
while True:
    input("To Throw a Dice, please click enter button") # If there is no input fun it automatically executes 1 to 100 entire fun
    face_val= random.randint(1,6)
    
    if count + face_val <=100:
        count+=face_val
    rolls+=1
    if count == 100:
        print(f"Your Dice value:{face_val}")
        print(f"You reached the Top in {rolls} chances")
        break
    print(f"Your Dice value:{face_val}")
    print(f"Your at {count} Position in {rolls} chances")
    print("*" * 50)
```
