### File Handling:
- The File is used to store the dat in electronic format (soft copy) in system
### File operation:
1. Opening a File -> open()
2. Reading a file -> read()
3. writing into a file -> write()
4. closing a file -> close()

#### 1.Opening a file:
- To Open a file by using open() function in that we need to pass file name and file opening mode
- syntax:
-        file_obj = open(file_name,mode)
-         # read/write operation
-         # close file
#### file opening modes:
1. r ->read --> opening file in read mode , if file not exist it raises file not found error
2. w ->write --> opening a file in write mode, if file not exist it creates a new file otherwise it erases the previous content
3. a ->append --> opening a file in append mode , if file not exist it creates a new file otherwise it adds the content to the previous or existing content
4. x ->create --> Used to crate a new file
5. b ->binary --> This mode is used to perform operations on binary files for writing we use wb mode for reading we use rb mode
6. r+ -> read + write
7. w+ -> write + read
8. a+ -> append + read

#### 2.Reading content from a file:
- To perform the read operation on a file for that we need to open the file in 'r' mode and then we can perform following operations
 1. read(): It returns entire file content as single string
 2. readline(): This method returns first line
 3. readlines(): It returns list of lines

#### 3.Writing into a file:
- If you open a file in write mode, we can perform following operations
  1. write(string): It add string into a file.
  2. writelines(List_of_line): It adds all list line to a file.

#### 4.Closing a file:
- After performing the read (or) write operations on a file, we need to close the file by using close function (or) close method otherwise data loss will happend.
