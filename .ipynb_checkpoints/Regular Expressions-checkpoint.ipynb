```python
import re
```

```python
# match 
string= "The sky is blue and ocean is blue"
res=re.match("The",string)
print(res)
```

```python
# match 
string= "The sky is blue and ocean is blue"
res=re.match("The",string)
print(res)
print(res.span()) #It returns the matched pattern positions
print(res.group()) #It returns the matched result
```

```python
# match(pattern,string): It checks whether the string is starts with pattern or not
try:
    string= "The sky is blue and ocean is blue"
    res=re.match("The",string)
    print(res)
    print(res.span()) #It returns the matched pattern positions
    print(res.group()) #It returns the matched result
except:
    print("There is no matched pattern")
```

```python
# search(pattern,string): It checks whether the string is starts with pattern or not
try:
    string= "The sky is blue and ocean is blue"
    res=re.search("sky",string)
    print(res)
    print(res.span()) #It returns the matched pattern positions
    print(res.group()) #It returns the matched result
except:
    print("There is no matched pattern")
```

```python
# serach(pattern,string): It checks whether the string is starts with pattern or not
try:
    string= "The sky is blue and ocean is blue"
    res=re.search("z",string)
    print(res)
    print(res.span()) #It returns the matched pattern positions
    print(res.group()) #It returns the matched result
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall("blue",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall("[abc]",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall("[abco]",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# split(pattern,string): It checks returns the all matched patterns as list
try:
    string= "The sky is blue and ocean is blue"
    res=re.split("blue",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# split(pattern,string): It checks returns the all matched patterns as list
try:
    string= "The sky is blue and ocean is blue"
    res=re.split("[abc]",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# sub()(old_pattern,new_pattern): It splits the string whenever the pattern is matched.
try:
    string= "The sky is blue and ocean is blue"
    res=re.sub("blue","red",string)
    print(res)
except:
    print("There is no matched pattern")
```

### meta characters

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall("[isabce]+",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall("[a-z]+",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall("[a-zA-Z]+",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue and ocean is blue"
    res=re.findall(r"[\D]+",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# findall(pattern,string): It returns the all matched patterns as list.
try:
    string= "The sky is blue123 and ocean8 is blue1234"
    res=re.findall("[0-9]+",string)
    print(res)
except:
    print("There is no matched pattern")
```

```python
# validate mobile number
try:
    mobile=input()
    res=re.fullmatch("^[6-9][0-9]{9}",mobile)
    print(res)
    if res:
        print(f"{mobile} is valid")
    else:
        print(f"{mobile} is invalid")
except:
    print("There is no matched pattern")
```

```python
# PANCARD number validation
try:
    pan_no=input()
    res=re.fullmatch("^[A-Z]{5}[0-9]{4}[A-Z]$",pan_no)
    print(res)
    if res:
        print(f"{pan_no} is valid")
    else:
        print(f"{pan_no} is invalid")
except:
    print("There is no matched pattern")
```

```python
# email validation
import re
try:
    email=input()
    res=re.fullmatch(r"^[a-z0-9][a-z0-9.]+\@[a-z0-9]+\.[a-z]{2,3}$",email)
    #print(res)
    if res:
        print(f"{email} is valid")
    else:
        print(f"{email} is invalid")
except:
    print("There is no matched pattern")
```

```python
# date validation
import re
try:
    date=input()
    res=re.fullmatch(r"^(0?[1-9]|[12][0-9]|3[01])[/-](0?[1-9]|1[0-2])[/-](\d{4})",date)
    #print(res)
    if res:
        print(f"{date} is valid")
    else:
        print(f"{date} is invalid")
except:
    print("There is no matched pattern")
```
