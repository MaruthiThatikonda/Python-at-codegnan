### Exception Handling 
- Exception is nothing but it is an error. whenever exception is occurs it stops the program execution and it returns type of exception and follwed by exception method.

#### Types of errors
- 1.Compiletime errors (or) syntax errors
- 2.Runtime errors
- 3.Logical errors.

#### 1.Compile (or) Syntax error.
- The syntax error raise during compilation time of the program, those errors call it self as syntax errors.
- EX: Indentation error, syntax error

#### 2.Runtime errors. / Runtime exception.
- The Runtime errrors raised during the runtime of the program. Those errors called as Runtime errors.
- EX: KeyError,valueError,ZeroDivisionError,IndexError,TypeError,etc,...
- We can't handle syntax errors, but we can handle the RuntimeErrors by using Exception Handling blocks(i.e, try-except)blocks.
  

#### Exception Handling Blocks
##### syntax:
 - try:
     - #block of code that code might contains exception.
 - except:
     - #exception handling block
 - else:
     - #this block executes when there is no exception occured in try block.
 - finally:
     - #this block always executes even the exception raised or not.

```python
#try block with exception
try:
    list=[1,2,3]
    print(list(10))
    print("hi")
except:
    print("something wrong in try block")
else:
    print("no exception raised in try block")
finally:
    print("Program done")
```

```python
try:
    list=[1,2,3]
    print(list.index(2))
    print("hi")
except:
    print("something wrong in try block")
else:
    print("no exception raised in try block")
finally:
    print("Program done")
```

```python
# Runtime errors
a=10
b=a.append(0)
print(b)
```

```python
# runtime errors
try:
    a=10
    b=a.append(0)
    print(b)
except Exception as e:
    print (f"something wrong in try block{e}")
```

```python
# runtime errors
try:
    a=10
    b=a.append(0)
    print(b)
except ZeroDivisionError as e:
    print(f"ZeroDivision not possible: {e}")
except ValueError as e:
    print(f" ValueError raised in try block: {e}")
    
#except Exception as E:
    #print (f"something wrong in try block{E}")
```

```python
# It executes the specific exception block when the specified exception raised otherwise, it executes the common except block
try:
    a=10
    b=a.append(1)
    c=a/b
    print(c)
except ZeroDivisionError as e:
    print(f"ZeroDivision not possible: {e}")
except ValueError as e:
    print(f" ValueError raised in try block: {e}")
    
except Exception as E: # common excepti0on block with Exception class
    print (f"something wrong in addition.add {E}")
```
