## STRING METHODS 
- Case conversion methods....
  - upper() : It converts entire string into Upper case 
  - lower() : It converts entire string into lower case
  - title() : It converts the first letter of every word into upper and remaining lower
  - swapcase(): It converts Upper case letter into lowercase and Lower to upper
  - capitalize(): It converts First character into upper case and remaining into lower case.

```python
string= "The Sky is blue and Ocean is blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

## REMOVE WHITE SPACE OR PATTERN AT ENDS OF STRING
  - strip() :By default it removes the white space at both ends of string
  - lstrip():It removes the white space at left side of string
  - rstrip():It removes the white space at right side of string

```python
s="      Pythonclass      "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@Python@@class@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
# split(): It splits the entire string into list of substring based on specified separetor
# by default it splicts at space 
string= "The Sky is blue and Ocean is blue"
print(string.split())
print(string.split("is"))
print(string.split("e"))
```

```python
#'sep'.join(iterable): It joins the all iterable elements with seperator
words = ['The', 'Sky', 'is', 'blue', 'and', 'Ocean', 'is', 'blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
```

# Substring Index finding
- index(substring) : If substring is exists in string, it returns the first occurencee of substring otherwise 'ValueError'
- find(substring)  : If substring is exists in string, it returns the first occurencee of substring position, otherwise -1

```python
string = "The Sky is blue and Ocean is blue"
print(string.index('is'))
```

```python
string = "The Sky is blue and Ocean is blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is',10))
print(string.find('blue',14))
print(string.find('e',-5))
```

```python
#count(substring) : It returns the total non -overlapping occurence of substring 
s= "ABCBCBCbbabab"
print(s.count("B"))
print(s.count("BCB"))
lower_string = s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('Z'))
```

```python
# replace (old_substring, new_substring): It replaces old sub-string with new sub-string 
string = "The Sky is blue and Ocean is blue"
print(string.replace("blue","Red"))
print(string.replace('e','E'))
print(string.replace('e','E',2)) # it replaces on the first 2 occurance
print(string.replace('e','E',-2))
```

### String Checking Methods 
 **String checking methods returns the Boolean value as result**
- isupper() :It return True, If entire string is in upper case
- islower() :It return True, If entire string is in lower case
- isalpha() :It return True, If entire string contains alphabets 
- isdigit() :It return True, If entire string contains digits
- isalnum() :It return True, If entire string contains either alphabets or digits
- startswith(substring): It returns True, If string starts with substring  
- endswith(substring): It return True, If string ends with substring

```python
print("CODE".isupper())
print("CODE".islower())
print("Code".islower())
print("1234".isdigit())
print("shreyas@96".isalnum())
print("96".isalpha())
print('code'.startswith('c'))
print("shryeas".endswith('s'))
```
