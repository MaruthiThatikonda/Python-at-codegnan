#### Class:
  - A class is a blue print of a real world object or template of object(Object is a instance of class).
  - Class is a collection of attributes (i.e, data and methods(behaviour))
  - syntax:
    - class class_name:
       - ##### attributes
            - ##### methods

#### Object:
   - Object is a instance of class and it is a real world item.
   - syntax:
       - object creation
          - obj=class_name(arguments)
             


#### Attributes :
  - The variables are defined inside the class those variables called as attributes.
  - The attributes are used for storing data.
  - #### Types
      - 1.Class Attributes
      - 2.Instance attributes.

```python
# Define student class
class student:
    school="XYZ" # class attributes
    def __init__ (self, name, age): #Initialization Constructor (Magic Method).  name and age parameters.
        self.name=name # instance attributes 
        self.age=age # instance attributes
       # print("This student class __init__ method")
    def details(self): #function
        return f"My name is {self.name} and Iam a {self.age} years old"
        
s1=student("Shreyas",31) 
s2=student("MSD",45)
print(s1.name)
print(s2.name) # attribute calling 
print(s1.details()) # method calling
print(s2.details())
print(s1)
print(s1.school)
print(s2.school)
```

### Methods: 
- The function defination is implement inside the class those functions called as methods.
 #### Types of Methods
          1. Instance Methods: 
          2. Class Methods : 
          3. Static Methods 

```python
class Student:
    school = "CODEGNAN School"

    def __init__(self, stdid, name, standed):
        self.name = name
        self.stid = stdid
        self.standed = standed

    def student_details(self):
        return f"My self {self.name} and I am from {self.standed} class"


s = Student(101, "Shreyas", 57)

print(s.name)
print(s.stid)
print(s.student_details())
```

#### Instance Method :  
- The method defination take 1st parameter as "self" those methods called as 'Instance Method'. This method works on instance attributes.
   - syntax:
     -  class class_name:           <br>
        - def method_name(self, parameters):     <br>
            -  block of code         <br>

#### Class Method :
  - The Method defination starts with 'classmethod' decorator and it takes 1st parameter as 'cls' those methods called as 'Class methods'. This methods works on class attributes.
  - syntax:
      -  class class_name:          <br>
         - @classmethod             <br>
        - def method_name(cls, parameters):     <br>
             -  block of code        <br>
    

#### static method:
- This Method defination starts with 'staticmethod' decorator and it doesnot take 1st parameter as 'self or cls' those methods called as 'Static methods'. This methods works on except instance attributes and class attributes.
   - syntax:
      -  class class_name:          <br>
         - @staticmethod             <br>
        - def method_name(parameter):    <br>
            -  block of code        <br>
  

```python
class student:
    school = "CODEGNAN School"  # class attribute

    def __init__(self, stdid, name, standed):
        self.name = name  # instance attributes
        self.stid = stdid  # instance attributes
        self.standed = standed   # instance attributes

    def student_details(self):   # instance attributes
        return f"My self {self.name} and I am from {self.standed} class"

    # class method
    @classmethod    # @ it is called decoretor
    def update_school(cls, new_school_name):
        cls.school=new_school_name
        return "School Name Updated"

    #static method
    @staticmethod
    def percentage(marks:list):
        return (sum(marks)/300) *100

s1= student(101,"Shreyas",57)
s2=student(102,"Mahi",10)
print(s1.school, s2.school)
print(s1.student_details())
print(s1.update_school("RAINBOW School"))
print(s1.school,s2.school)
print(s2.percentage([85,68,95]))
```

### Encapsulation:
- Encapsulation means , wraping(keeping in single unit) the method and attributes in single unit that one is class. The main aim of encapuslation is to restrict the direct acces of attributes and methods.
- These type of Restriction can implement by using 'Access modifiers'.
- Types of Access modifiers:
     - 1. Public
     - 2. Protected
     - 3. Private
- In python we dont have any keywords to represent it an public, protected and private but by using underscore (_) we can define.
   #### 1. Public
   - The public members(i.e, attributes + methods) doesn't starts with underscore. This public members can access anywhere in the program.
   #### 2. Protected
   - The Protected members starts with singleunderscore(_) . This members can access in inside the class and in sub classes also.
   #### 3. Private
   - The private members starts with double underscore(__). This members can access in only inside.

```python
class Bank:
    def __init__(self,name,account,password,balance):
        self.name=name
        self.acc=account
        self.__password=password  # __ private
        self._bal=balance  # protected
    def __total_users(self): # getter method
        return "Total users: 101"
    def get_balance(self): # getter method
        return f"Current balance is {self._bal}"
    def deposite(self, deposite_amount): # setter method
        self._bal += deposite_amount
        return f"{deposite_amount} Deposited Successfully"

u1=Bank("Msd",1001,"msd@123",10000)
u2=Bank("Gill",1002,"gill@123",20000)
print(u1.name)
print(u1.get_balance())
print(u1.deposite(5000))
print(u1.get_balance())
print(u1._bal) # protected members access in outside the class
print(u1._Bank__total_users()) # Name mangling: to acess private members also  (syntax: obj._classname_attribute)
```
