### Tuple
 - Tuple is ordered collection of different data type elements .Tuple is immutable and defines with '()'
 - Characteristics:
    - Tuple is oredered collection
    - Tuple is immutable
    - Tuple is allows indexing slicing
    - Tuple is Heterogenous datatype(allows diff datatypes of elts)
    - Tuple allows duplicate items.
 - Ex: t=(1,2,3.5,'Hello',[10,20,30])
