### SETS
 - Sets is unordered collection of different datatype elements sets are mutable and defines with '{ }'

## characterstics 
 - Set is a unordered collection
 - set elements are unique
 - set doesnot supports indexing
 - set are mutable
 - set allows hashable datatype as elements
 - set elements can be used as dictionary keys


```python
#set unorder collection
# set return unique elements
s={1,2,3,2,3,'hi','a',2.02,(10,20,30)} 
print(s)
#print(s[0]) # TypeError: 'set' object is not support indexing 
```

```python
print({10,6,3,5})
print(hash(10),hash(3))
```

```python
# List can't be used as set element
s={1,2,3,2,3,'hi','a',2.02,(10,20,30)}
print(s)
```

```python
# empty set representation
s1={ }
s2=set()
print(type(s1), type(s2))
```
