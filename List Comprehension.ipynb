### List Comprehension:
 - This is the one of way to create a list with an expression.
 #####  syntax:
     -[expression for iterator in iterable if condition]. 

```python
# generate 1-n numbers by using list comprehension
n=int(input())
res=[x for x in range(1,n+1)]
print(res)
```

```python
# first n even numbers by using list comprehension
n=int(input())
res=[x*2 for x in range(0,n)]
print(res)
```

```python
# first n quibic numbers by using list comprehension
n=int(input())
res=[x**3 for x in range(1,n+1)]
print(res)
```

```python
# find the odd numbers between n1 to n2
n1,n2=map(int,input().split())
res=[x for x in range(n1,n2+1) if x%2==1] # we can use x%2!=0
print(res)
```

```python
# in given arr of integers perform the following operations
# if even -find the square of a number
# if odd -find squareroot value upto 2 decimals
arr=list(map(int,input().split()))
res=[x**2 if x%2==0 else round(x**0.5,2) for x in arr]
print(res)
```

```python
# dice combinations
res=[(x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
# return the dice combination if sum of two face values is divisble by 3
res=[(x,y) for x in range(1,7) for y in range(1,7) if (x+y)%3==0]
print(res)
```

```python
# return any one dice face value is 4 
res=[(x,y) for x in range(1,7) for y in range(1,7) if x==4 or y==4]
print(res)
```

```python
# count the number of vowels exists in string
string =input()
res=[char for char in string.lower() if char in "aeiou"]
print(res)
```

```python
# set comprehension 
# 1 to n numbers
n=int(input())
res={ x for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
# dictionary comprehension 
# 1 to n numbers
n=int(input())
res={ x:x**2 for x in range(1,n+1)}
print(res)
print(type(res))
```

```python
# tuple comprehension 
# 1 to n numbers
n=int(input())
res=(x for x in range(1,n+1))
print(res)
print(type(res))
# generator is short cut 
```

```python
# tuple comprehension 
# 1 to n numbers
n=int(input())
res=tuple (x for x in range(1,n+1))
print(type(res))
print(res)
```

#### assert keyword:
   - This keyword is used for testing

```python
age=int(input())
if age>=18:
    print("Eligible")
else:
    print("Not eligible")
```

```python
# assert keyword : This keyword is used for testing
age=int(input())
assert age>=18,"Age must be above 17"
print("Eligible")
```

```python
# assert keyword : This keyword is used for testing
age=int(input())
assert age>=18,"Age must be above 17" # rises AssertionError: Age must be above 17
print("Eligible")
```

```python
arr=list(map(int,input().split()))
first=second=0
for num in arr:
    if num>first:
        first=num
    elif first>num>second:
        second=num
print("Second largest elt:",second)
```

```python
# reverse of array
arr = list(map(int, input().split()))

l = 0
r = len(arr) - 1

while l < r:
    arr[l], arr[r] = arr[r], arr[l]
    l += 1
    r -= 1

print(arr)
```

```python
# rotate array by one position
arr = list(map(int, input().split()))
last=arr[-1]
for i in range(len(arr)-1,0,-1):
    arr[i]=arr[i-1]
arr[0] = last
print(arr)
```
