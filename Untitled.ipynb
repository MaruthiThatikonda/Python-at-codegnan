```python
#slicing
#str_obj[start:end:step]
s="Maruthi"
print(s[1:5:1])
print(s[5:])
print(s[-4:])
print(s[::-1])
print(s[-3::1])
```

```python
#empty string Representation
#type(): This fun returns the objects datatype
s1 =""
s2=str()
print(type(s1))
print(type(s2))
```

```python
#string reading from user
#input() : This fun is used to read dynamically from user and 
#          This fun by default read it as string data type
s=input()
print(type(s))
print(s)
```

```python
#string operations 
# 1.string concatination (using + operator)
# 2.string Representation( using * operator)
s1="MS"
s2="Dhoni"
print(s1+s2)
s3=s1+" "+s2
print(s3)
print(s3 * 3)
```
