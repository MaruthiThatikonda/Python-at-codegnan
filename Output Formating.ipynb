### Output Formatting
 - 1.old style format(using commass)
 - 2. Accesss specifier format
 - 3. Dot format
 - 4. F format

```python
#1. Normal old format
name=input()
age=int(input())
print("My name is",name,"and Iam",age,"years old")
```

```python
# using string concatination
name="msd"
age=40
print("My name is "+name+" and Iam "+str(age)+" years old")
```

```python
# 2.Access Specifier
#string- %s
#integer- %d
#float- %f
name="msd"
age=40
print("My name is %s and Iam %d years old"%(name,age))
```

```python
name="msd"
age=40
print("My name is %s and Iam %d years old"%(name,age))
```

```python
# 3. dot formatting
name="msd"
age=40
print("My name is {} and Iam {}years old".format(name,age))
print("My name is {} and Iam {}years old".format(age,name))
```

```python
# 4.f format 
name="Msd"
age=40
print(f"My name is {name} and Iam {age} years old")
print(f"My name is {age} and Iam {name} years old")
```
