### Dictionary
- Dictionary is ordered collection key-value (ie, items) pairs. It is mutable and defines with '{ }'
- Dictionary form:
   - { key:value}
 
#Characterstics:
  - Dictionary is ordered collection
  - Dictionary is mutable
  - Dictionary doesnt support indexing
  - Dictionary keys are unique
  - Dictionary values allows duplicate
  - Dictionary values can be accesed by using keys 
  - List cant be used as dictionary key, because unhashable
      

```python
Details={"Name":"Shreyas","Course":"Python","Batch":57,"Age":22}
print(Details)
```

```python
#Dictionary values access by using Keys
#updating dictionary values 
#adding new items to dictionary
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
print(details['course'])
# print(details['location']) # KeyError: 'location'
#updating dictionary values
details['age']=25
print(details['age'])
#adding new items
details['location']="Hyderabad"
print(details)
# to add age in exsiting value
details['age']=details['age']+1
```

### Dictionary methods
- get(key,default_vlaue)(optional of default value): It returns the value if key is exists in dictionary, otherwise it returns default value. 
- update(new_dictionary): This method updates the values if new _dictionary keys already exists in Dictionary, otherwise it adds new items to dictionary

```python
# get method
#dictionary keys are unique
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
print(details.get('name'))
print(details.get('marks')) # it returns none,by default bcz key not exist in details
print(details.get('marks','marks not avaliable')) ##it returns default value. 
```

```python
# update method
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
details.update({'batch':58,'course':'Java','Location':"HYD"}) #add location in existing details and remaing will update values
print(details)
```

  - keys(): This method returns the list dictionary keys
  - values() : This method returns the list dictionary values
  - items(): This method return the list of Tuple of key-value pairs 

```python
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
print(details.keys())
print(details.values())
print(details.items())
```

- pop item:It returns and removes the last item as tuple
- pop(key):This method removes the specified key item and return the specified key's value,otherwise "KeyError"

```python
details={"name":"Shreyas","course":"Python","batch":57,"age":22}
item=details.popitem()
print(item)
print(details)
val=details.pop('name')
print(val)
print(details)
```
