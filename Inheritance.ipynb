### Types of Inheritance

```python
# types of inheritance
# single inheritance
# without child class implementation
# when child class dont have constructor in that case the parent class constructor behave likes child class constructor.
class person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and I am {self.age} years old."

class student(person):
    pass

s = student("Shreyas", 30, "Male")
print(s.name)
print(s.details())
```

```python

# 1.single inheritance
#  child class with constructor
# when child class dont have constructor in that case the parent class constructor behave likes child class constructor.
class person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and I am {self.age} years old Gender {self.gender}"

class student(person):
    def __init__(self, stdid, name, age,  branch):
        self.name=name
        self.age=age
        self.stdid=stdid
        self.branch=branch
    

s = student(101,"Shreyas", 30, "CSM")
print(s.name)
print(s.branch)
print(s.gender) # AttributeError: 'student' object has no attribute 'gender'
print(s.details()) # it raises attribute error because student reference doesnot contains the gender attribute.
```

```python
# by using of super() class.
class person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and I am {self.age} years old Gender {self.gender}"

class student(person):
    def __init__(self, stdid, name,gender, age,  branch):
        super().__init__(name,age,gender)
        self.stdid=stdid
        self.branch=branch
    

s = student(101,"Shreyas","male",30, "CSM")
print(s.name)
print(s.branch)
print(s.gender) 
print(s.details())
```

```python
# by using of super() function. (sir)
class person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and I am {self.age} years old Gender {self.gender}"

class student(person):
    def __init__(self, stdid, name,gender, age,  branch):
        super().__init__(name,age,gender)  # inherit the properties of name age gender from person class.
        self.stdid=stdid
        self.branch=branch
    def percentage(self,marks):
        return sum(marks)/3
    

s = student(101,"Shreyas","male",30, "CSM")
print(s.name)
print(s.branch)
print(s.gender) 
print(s.details())
print(s.percentage([82,55,85]))
```

```python
# 2.Multilevel Inheritance

class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def details(self):
        return f"My name is {self.name} and I am {self.age} years old. Gender: {self.gender}"


class Student(Person):
    def __init__(self, stdid, name, age, gender, branch):
        super().__init__(name, age, gender)
        self.stdid = stdid
        self.branch = branch

    def student_details(self):
        return f"Student ID: {self.stdid}, Branch: {self.branch}"


class Employee(Student):
    def __init__(self, stdid, name, age, gender, branch, empid, role):
        super().__init__(stdid, name, age, gender, branch)
        self.empid = empid
        self.role = role

    def employee_details(self):
        return f"Employee ID: {self.empid}, Role: {self.role}"

e = Employee(101, "Mahi", 20, "Male", "AI", 123, "Software Engineer")

print(e.name)
print(e.age)
print(e.gender)
print(e.stdid)
print(e.branch)
print(e.empid)
print(e.role)
print(e.details())
print(e.student_details())
print(e.employee_details())
```

```python
# 3.Multiple Inheritance

class Father:
    def __init__(self, father_name):
        self.father_name = father_name

    def father_details(self):
        return f"Father Name: {self.father_name}"

class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name

    def mother_details(self):
        return f"Mother Name: {self.mother_name}"

class Child(Father, Mother):
    def __init__(self, father_name, mother_name, child_name):
        Father.__init__(self, father_name)
        Mother.__init__(self, mother_name)

        self.child_name = child_name

    def child_details(self):
        return f"Child Name: {self.child_name}"


c = Child("Ramesh", "Sita", "Rahul")

print(c.father_details())
print(c.mother_details())
print(c.child_details())
```

```python
class father:
    def __init__(self, fathername):
        self.fathername = fathername

    def work(self):
        return "pump op"


class mother:
    def __init__(self, mothername):
        self.mothername = mothername

    def work(self):
        return "house wife"


class son(father, mother):
    def __init__(self, name, age, fathername, mothername):
        father.__init__(self, fathername)
        mother.__init__(self, mothername)
        self.name = name
        self.age = age

    def parents_works(self):
        print(father(self.fathername).work())
        print(mother(self.mothername).work())


s = son("msd", 22, "Pan singh", "devika")

print(s.work())
print(son.mro())
s.parents_works()
```

```python
#4. HierachialInheritance
# Parent class
class Father:
    def __init__(self, fathername):
        self.fathername = fathername

    def work(self):
        return "Pump Operator"


# Child class 1
class Son(Father):
    def __init__(self, name, age, fathername):
        Father.__init__(self, fathername)
        self.name = name
        self.age = age

    def details(self):
        print("Son Name :", self.name)
        print("Age      :", self.age)
        print("Father   :", self.fathername)
        print("Father Work :", Father(self.fathername).work())


# Child class 2
class Daughter(Father):
    def __init__(self, name, age, fathername):
        Father.__init__(self, fathername)
        self.name = name
        self.age = age

    def details(self):
        print("Daughter Name :", self.name)
        print("Age           :", self.age)
        print("Father        :", self.fathername)
        print("Father Work   :", Father(self.fathername).work())


# Objects
s = Son("MSD", 22, "Pan Singh")
d = Daughter("Anu", 20, "Pan Singh")

print(s.work())
print(Son.mro())
s.details()

print()

print(d.work())
print(Daughter.mro())
d.details()
```

```python
# 5.Hybrid Inheritance
# Parent class
class Father:
    def __init__(self, fathername):
        self.fathername = fathername

    def father_work(self):
        return "Pump Operator"


# Parent class
class Mother:
    def __init__(self, mothername):
        self.mothername = mothername

    def mother_work(self):
        return "House Wife"


# Child class (Hierarchical)
class Son(Father):
    def __init__(self, name, age, fathername):
        Father.__init__(self, fathername)
        self.name = name
        self.age = age

    def son_details(self):
        print("Son Name :", self.name)
        print("Age :", self.age)
        print("Father :", self.fathername)
        print("Father Work :", Father(self.fathername).father_work())


# Child class (Hierarchical)
class Daughter(Father):
    def __init__(self, name, age, fathername):
        Father.__init__(self, fathername)
        self.name = name
        self.age = age

    def daughter_details(self):
        print("Daughter Name :", self.name)
        print("Age :", self.age)
        print("Father :", self.fathername)
        print("Father Work :", Father(self.fathername).father_work())


# Grandchild class (Multiple Inheritance)
class GrandSon(Son, Mother):
    def __init__(self, name, age, fathername, mothername):
        Son.__init__(self, name, age, fathername)
        Mother.__init__(self, mothername)

    def family_details(self):
        print("Name :", self.name)
        print("Age :", self.age)
        print("Father :", self.fathername)
        print("Mother :", self.mothername)
        print("Father Work :", Father(self.fathername).father_work())
        print("Mother Work :", Mother(self.mothername).mother_work())


# Object
g = GrandSon("MSD", 22, "Pan Singh", "Devika")

print(GrandSon.mro())
g.family_details()
```
